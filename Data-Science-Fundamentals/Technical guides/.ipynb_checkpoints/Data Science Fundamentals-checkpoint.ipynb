```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv("gas_prices.csv")
gas.head()
```

```python
plt.figure(figsize=(5,3),dpi=330)
plt.title("Gas price growing in Europe countries")
plt.plot(gas.Year,gas.France,"b.-")
plt.plot(gas.Year,gas.Canada,"g.-")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5,3),dpi=330)
# plt.plot(gas.Year,gas.France,"b.-")
# plt.plot(gas.Year,gas.Canada,"g.-")
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker='.')
        
plt.title("Gas price growing in Europe countries")
plt.legend()
plt.show()
```
